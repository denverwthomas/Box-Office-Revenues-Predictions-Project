```python
def clean_df(df):
    header=df.iloc[7]
    df=df[10:]
    df=df.rename(columns = header)
    df=df.rename(columns={df.columns[0]:'DATE'})
    df['DATE']=pd.to_datetime(df['DATE'])
    df=df.set_index('DATE')
    return df

def clean_quarter_df(df):
    header=df.iloc[7]
    df=df[10:]
    df=df.rename(columns = header)
    df=df.rename(columns={df.columns[0]:'DATE'})
    df['DATE']=pd.to_datetime(df['DATE'].str.split(' ').apply(lambda x:''.join(x[::-1])))

    df=df.set_index('DATE')
    return df

def rename_to_df(prev,df):
    column_len=len(df.columns)
    for i in range( column_len):
        current_name=df.columns[i]
        new_name=prev+current_name
        df=df.rename(columns={df.columns[i]:new_name})
    return df
```

```python
# China GDP 
import pandas as pd
df_cGDP=pd.read_csv("Econ_data/china GDP.csv")
```

```python
# China GDP 
import pandas as pd
df_cGDP=pd.read_csv('Econ_data/china GDP.csv')
df_cGDP=clean_quarter_df(df_cGDP)
df_cGDP=df_cGDP.loc[:,:'Real GDP']
df_cGDP=rename_to_df('China: ',df_cGDP)
# China expenditure
import pandas as pd
df_c_personal_exp=pd.read_csv('Econ_data/china personal expenditure.csv')
df_c_personal_exp=clean_df(df_c_personal_exp)
df_c_personal_exp=rename_to_df('China Personal Expense: ',df_c_personal_exp)
# US CPI
import pandas as pd
df_CPI=pd.read_csv('Econ_data/CPI 1970.csv')
df_CPI=clean_df(df_CPI)
df_CPI=rename_to_df('US CPI: ',df_CPI)
#gini coef
import pandas as pd
df_gini=pd.read_csv('Econ_data/gini coefficient.csv')
df_gini=clean_df(df_gini)
df_gini=rename_to_df('Gini: ',df_gini)
#US Real GDP
import pandas as pd
df_R_GDP=pd.read_csv('Econ_data/US real GDP.csv')
df_R_GDP=clean_quarter_df(df_R_GDP)
df_R_GDP=rename_to_df('US Real GDP: ',df_R_GDP)
#US nominal GDP
import pandas as pd
df_N_GDP=pd.read_csv('Econ_data/US nominal GDP.csv')
df_N_GDP=clean_quarter_df(df_N_GDP)
df_N_GDP=rename_to_df('US Nominal GDP: ',df_N_GDP)
#World GDP
import pandas as pd
df_W_GDP=pd.read_csv('Econ_data/world GDP 2000-2019.csv')
df_W_GDP=clean_quarter_df(df_W_GDP)

df_W_GDP=rename_to_df('World GDP: ',df_W_GDP)
#World CPI
import pandas as pd
df_W_CPI=df_nomial_GDP=pd.read_csv('Econ_data/World CPI.csv')
df_W_CPI=clean_quarter_df(df_W_CPI)
df_W_CPI=df_W_CPI.rename(columns={df_W_CPI.columns[0]:'World CPI 1'})
df_W_CPI=df_W_CPI.rename(columns={df_W_CPI.columns[1]:'World CPI 2'})
```

```python
import pandas as pd
df1=pd.read_csv('Econ_data/deflator.csv')
df2=pd.read_csv('Econ_data/All-Transactions House Price Index .csv')
df3=pd.read_csv('Econ_data/PPIACO.csv')
df4=pd.read_csv('Econ_data/Real GDP.csv')
df5=pd.read_csv('Econ_data/Unemployment Rate.csv')

df1['DATE']=pd.to_datetime(df1['DATE'])
df2['DATE']=pd.to_datetime(df2['DATE'])
df3['DATE']=pd.to_datetime(df3['DATE'])
df4['DATE']=pd.to_datetime(df4['DATE'])
df5['DATE']=pd.to_datetime(df5['DATE'])

df1=df1.set_index('DATE')
df1=df1.rename(columns={df1.columns[0]:'GDP_deflator'})
df2=df2.set_index('DATE')
df2=df2.rename(columns={df2.columns[0]:'House_Price_Index'})
df3=df3.set_index('DATE')
df3=df3.rename(columns={df3.columns[0]:'Producer_Price_Index'})
df4=df4.set_index('DATE')
df4=df4.rename(columns={df4.columns[0]:'US Real GDP 2'})
df5=df5.set_index('DATE')
df5=df5.rename(columns={df5.columns[0]:'Unemployment_Rate'})

df1.index=df1.index.to_period("Q")
df2.index=df2.index.to_period("Q")
df3.index=df3.index.to_period("M")
df4.index=df4.index.to_period("Q")
df5.index=df5.index.to_period("M")
```

#### 3.1 Breakdown Releasedate into M/Q/Y  and Merge!

```python
df['release date']=pd.to_datetime(df['release date'])
df['release month']=df['release date'].dt.to_period("M")
df['release quarter']=df['release date'].dt.to_period("Q")
df['release year']=df['release date'].dt.to_period("Y")

df=df.merge(M_Index_df, left_on='release month', right_on='DATE')
df=df.merge(Q_Index_df, left_on='release quarter', right_on='DATE')
df=df.merge(Y_Index_df, left_on='release year', right_on='DATE')
```
